```python
#there are two basic object types in pandas that hold data and one is called
# the DATAFRAME and it is basically just a table of rows and columns and then each of those
# columns is known as a pandas series.

# YOU CAN HAVE A PANDA SERIES THAT IS NOT PART OF A DATAFRAME but mostly
# we work with series that are part of a data frame.

#The question is how do I select a series from it perhaps to do an analysis 
#on a particular series or to manipulate that series 

```

```python
import pandas as pd
```

```python
# We will use the data from UFO reports from 1930 to 2000
# read csv use the ',' as default separator.

ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
# to check the data
ufo
```

```python
# we can confirm that this is a DataFrame

type(ufo)
```

```python
ufo.head()
```

```python
# How we do select a series? 
# PRETTY IMPORTANT TO USE THE CAPITAL lETTERS otherwise we get an error
# Basic way is to use is as BRACKET NOTATION: 

ufo['City'] # in this case we will have the series just for 'City'

```

```python
# to confirm that is a series we can type:

type(ufo['City'])
```

```python
# we can use a shortcut for the same calling series example ufo['City']

ufo.City # we will get the same information as the one mention before
```

```python
# If you have in your data a column like the one in this data "Color Reported" 
# that is using space to separete the two words, it is NOT possible the use 
# shortcut, we need to use the BRACKET NOTATION

ufo['Colors Reported']
```

```python
ufo.shape
```

```python
# IMPORTANT DOT NOTATION ONLY WORKS UNDER CENTAIN CIRCUMSTANCES WHEREAS 
# BRACKET NOTATIION WILL ALWAYS WORK
```

```python
# How do I create a new pandas series in a DataFrame

# In regular python 
'ab' + 'cd'
 
# the result will be a concatenation of abcd    
```

```python
# if we try this example with our data will be:

ufo.City + ufo.State

# we will get the name of the city + the state together:
```

```python
# we can do the same using a ',' and that way we have a more organize information

ufo.City + ', ' + ufo.State

# remenber to type space after the ', ' so you get the right separation
```

```python
# To create a new column with the information we have created before
# ufo.City + ', ' + ufo.State we have to do the following: 

# first give a name to the new column 

# for example: Location WE NEED to USE BRACKET NOTATION ufo['Location']  

# this new column should be directed to the one with the information that we want
# to show, in this case: ufo.City + ', ' + ufo.State

# if we try to use shortcut ufo.location = ufo.City + ', ' + ufo.State it WILL NOT work.

# The right way to do it is:

ufo['Location'] = ufo.City + ', ' + ufo.State
```
