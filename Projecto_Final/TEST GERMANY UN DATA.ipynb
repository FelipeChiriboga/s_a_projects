# LINEAR REGRESSION MODEL - TEST 1 COUNTRY

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df_Germany['avg_year'].values
y = df_Germany['Age_male'].values
```

```python
mlr = LinearRegression()
mlr
```

```python
X.shape, y.shape
```

```python
X = np.array(X).reshape(-1,1)
```

```python
y = np.array(y)
```

```python
X.shape, y.shape
```

```python
mlr.fit(X,y)
```

```python
mlr.coef_
```

```python
mlr.intercept_
```

```python
ypred = mlr.predict(X)
ypred
```

```python
plt.figure()
plt.plot(X, y, 'bo', label='Years')
plt.plot(X, ypred, 'r-', label ='Age per year')
plt.xlabel('avg_year')
plt.ylabel('Age_male')
plt.legend()
plt.title('Life Expectancy Germany')
plt.show()
```
