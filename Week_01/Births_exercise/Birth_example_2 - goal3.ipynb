```python

birthnames = []
a = 1880
while a <= 2017:
    fl = str("names/yob" + str(a) + ".txt")
    bth = pd.read_csv(fl, names=['names', 'sex','birthcnt']) #probably need to add the name column
    bth['year'] = a 
    birthnames.append(bth)
    a = a + 1
    
birthnames = pd.concat(birthnames)
birthnames
```

```python
birthnames.groupby('names') ['birthcnt'].sum().head().plot.bar()
```

```python
def get_initial(s):
    return s[0]

birthnames['initial'] = birthnames['names'].apply(get_initial)
birthnames
```

```python
#easy: create a bar plot with 5 names in one year.

import pandas as pd

df = pd.read_csv('names/yob1983.txt', names=['name','sex', 'birthcnt'])

```

```python
df
```

```python
df['birthcnt'].sum()
```

```python
df.head()
```

```python
def get_initial(s):
    return s[0]

df['initial'] = df['name'].apply(get_initial)
df
```

```python
%matplotlib inline
```

```python
df.groupby('initial') ['birthcnt'].sum().plot.bar()
```

```python
ini = df.groupby(['sex', 'initial'])['birthcnt'].sum()
topchars = ini.sort_values(ascending=False).head(10)
topchars.plot.bar()
```

```python
ini = df.groupby(['sex', 'name'])['birthcnt'].sum()
topchars = ini.sort_values(ascending=False).head(5)
topchars.plot.bar()
```
