```python
# Calculate the average age for all passengers. Discard passengers with missing age. 

# from the table we take the column "Age" calculate the mean and add an argument (Skip NAN) meaning no values.

avg_age = df.Age.mean(skipna=True)
avg_age
```

```python
# Calculate the average age for survived and drowned passengers separately.

# Survived (sr) average just for the people that survived, and then check the mean. 

sr = df[(df['Survived'] == 1)]
sr.Age.mean()
```

```python
# Drowned (dr) average just the dead people age average.

dr = df[(df['Survived'] == 0)]
dr.Age.mean()
```

```python
# Replace missing age values by the mean age.

''' 
to replace missing values (NaN) with average we have to use the function "fillna" meaning all the data that has NaN will be 
replaced with the value of the average age and inplace=True means that it will be always replaced as a TRUE value!
'''

df.Age.fillna(value=avg_age, inplace=True)
df.Age
```
