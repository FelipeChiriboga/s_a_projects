# The Fizz-Buzz Test

# Go through the numbers from 1 to 100:

1. If the number is divisible by 3, write "Fizz"
2. If the number is divisible by 5, write "Buzz"
3. If the number is divisible by 3 and 5, write "FizzBuzz"

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz', i)
    elif  i % 3 == 0:
        print('Fizz', i)
    elif i % 5 == 0:
        print('Buzz', i)

    
```
