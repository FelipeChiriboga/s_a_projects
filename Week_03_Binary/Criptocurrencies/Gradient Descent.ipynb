```python
def loss(a):
    return a ** 2 - 3 * a
```

```python
def derivative(a):
    return 2 * a - 3
```

```python
a = 42 
learning_rate = 0.3
max_iter = 100
threshold = 0.0001
```

```python
while max_iter:
    gradient = derivative(a)
    a = a - gradient * learning_rate
    max_iter -= 1
    print(f"{max_iter:6} {a:12.4f} {gradient:12.6f}")
    if abs(gradient) < threshold:
        print('CONVERGED!')
        break
```
