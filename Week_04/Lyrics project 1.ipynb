# 2. extract all links from the song list
   
   !! find a pattern by reading the HTML code
   
   string functions 
   
   BeautifulSoup
   
   Regular Expressions


```python

def myFunc(s):
   if s.startswith('a href="/lyric/'):# and s.endswith('" target="_blank"'):
       return True

links = list(filter(myFunc, split_1))
```

```python
# REGULAR EXPRESSIONS!!

import re

# <a href=\"([^\"]+)\"

html

pattern = r'href=\"(/lyric[/+\w]+)'
```

```python
sub_link = re.findall(pattern, html)
sub_link
```

```python
new_link = []

for i in sub_link:
    base = "https://www.lyrics.com"
    url = base + i
    new_link.append(url)

#    https://www.lyrics.com/lyric/5407228/Def+Leppard/Good+Morning+Freedom
    
```

# Download each song and write it to a HTML file

  !! try this for 3 songs first
  
  !! MUST include the lines inside the for loop:
  
  import time
  
  time.sleep(30)

```python
import time

artist = 'Def_Leppard'
n = 0

for url in new_links:
    
    result = requests.get(url)
    name = url.split('/')[-1]
    name = name.replace('+', '_')
    f = open(f'lyrics_{artist}-{name}.txt', 'w', encoding="utf=8")
    
    f.write(result.text)
    f.close()
    n += 1

    time.sleep(30)
```

```python
# new_link.append(url) - example for appending!!!
songs = [] # tokens

for file in dirs: 
    f=open(path + file, 'r', encoding="utf=8")
    stx = f.read() # stx is the foramt text file!!!!!
    f.close()
    
    lyrics = re.findall('data-lang="en">(.+)</pre>', stx, re.DOTALL)
    if len(lyrics) == 1:
        lyrics = lyrics[0]
        l2 = re.sub("\.|\,|\?|\n|<br>", " ", lyrics)
        l2 = re.sub("<[^>]+>", "", l2)
        l2 = l2.lower()
        tokens = l2.split()
        songs.append(tokens)
    else:
        print("something went wrong")
```
