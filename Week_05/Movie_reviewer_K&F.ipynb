```python
import pandas as pd

PATH = "movie_recommendations.xlsx"

df = pd.read_excel(PATH)
```

```python
print("""
    Kristians and Felipes Movie Recommender:

    Enter your favourite genre or reviwer
    and the program will recommend a movie.

""")
```

```python
print("Enter a rating number 1 - 5 (or leave it empty): ", end="")
rating = float(input())

if rating:
    r = df[df['Rating'] == rating]
else:
    r = df
    
print("Enter a reviewer (or leave it empty): ", end="")
reviewer = input()

if reviewer:
    re = r[r['Reviewer'] == reviewer]
else:
    re = r

print("Enter a genre (or leave it empty): ", end="")
genre = input()

if genre:
    g = re[re['Genre'] == genre]
else:
    g = re


if g.shape[0] > 0:
    result = g.sort_values('Rating', ascending=False)
    print(result.head(3).sample(1))
else:
    print("Sorry, no results!")
```

# 2 part DB connection 

```python
import sqlite3
```

```python
DB_SETUP = '''
CREATE TABLE IF NOT EXISTS 'reviews' (
title VARCHAR(250),
genre VARCHAR(64),
rating INTEGER,
reviewer VARCHAR(64)
);
''' 
```

```python
db = sqlite3.connect('movies.db')
db.executescript(DB_SETUP)
```

```python
for i, row in df.iterrows():
    print(row['Genre'], row['Rating'])
    query = 'INSERT INTO reviews VALUES (?,?,?,?)'
    db.execute(query,(row['Name'], row['Genre'], row['Rating'], row['Reviewer']))
    
    db.commit()
db.close()
```
