```python
q("SELECT genre, title, reviewer, rating FROM reviews WHERE reviewer='Helena' GROUP BY genre ORDER BY rating DESC, title ASC LIMIT 5")
```

```python
q("SELECT genre, count(title), avg(rating) FROM reviews GROUP BY genre HAVING count(title) >= 3 LIMIT 5")
```

```python
q("SELECT genre, count(title), avg(rating) FROM reviews GROUP BY genre HAVING count(title) >= 3 ORDER BY count(title) DESC LIMIT 10")
```
